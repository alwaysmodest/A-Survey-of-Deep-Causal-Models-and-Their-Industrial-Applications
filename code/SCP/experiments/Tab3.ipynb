```python
import pandas as pds
```

```python
dat = pds.read_csv('results2021/real_results.txt', sep=' ', header=None)
```

```python
n_sample = [int(x[1]) for x in dat[1].str.split('_')]
dat['level'] = n_sample
```

```python
dat.head()
```

```python
rename_dict = {
    'dor': 'NN', 'tarnet': 'CFR', 'cfrnet': 'CFR', 'propensity': 'NN-IPW', 'vsr': 'VSR', 'scp': 'SCP', 'drcrn': 'DR-CFR', 'bmc':'DEC', 'overlap': 'OP'
}
```

```python
names = [rename_dict[x] for x in dat[0]]
```

```python
dat['method'] = names
```

```python
dat['Rank'] = [f"{x:.1f}" + ' ({})'.format(str(round(y, 1))) for x, y in zip(dat[6], dat[7])]
```

```python
dat['RMSE'] = [f"{x:.3f}" + ' ({})'.format(str(round(y, 4))) for x, y in zip(dat[10], dat[11])]
```

```python
df1 = dat.pivot(index='method', columns='level', values='RMSE')
df1
```

```python
df2 = dat.pivot(index='method', columns='level', values='Rank')
df2
```

```python
df1.to_csv('Tab3_A.csv')
df2.to_csv('Tab3_B.csv')
```
